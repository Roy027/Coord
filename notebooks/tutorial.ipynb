## 3. Geometric Feature Engineering
We transform raw 3D coordinates into "Plane Normal Dot Products" which act as a symmetry fingerprint.

```python
# Sample Tetrahedral (T-4) coordinates
sample_xyz = [
    ("Center", (0, 0, 0)),
    ("L1", (1, 1, 1)),
    ("L2", (-1, -1, 1)),
    ("L3", (1, -1, -1)),
    ("L4", (-1, 1, -1))
]

# 1. Convert coordinates to plane normal vectors
planes = coordinates_to_planes(sample_xyz)
print(f"Generated {len(planes)} planes.")

# 2. Convert planes to dot product descriptors
descriptors = planes_to_dot_products(planes)
print(f"Descriptors (first 5): {descriptors[:5]}")
```

## 4. Machine Learning Classification
We use the `CoordinationClassifier` to predict the geometry from the descriptors.

```python
# Initialize classifier
clf = CoordinationClassifier(hidden_layer_sizes=(16,))

# In a real scenario, you would train on the provided Excel data:
# df = pd.read_excel("../data/training/trainingdata.xlsx")
# clf.train(X_train, y_labels)

# For this demo, we use a mock prediction setup
mock_features = np.random.rand(1, 15) # Example for CN4 (comb(6,2)=15 dots)
try:
    prediction, probabilities = clf.predict(mock_features)
    print(f"Predicted Geometry: {prediction[0]}")
    print(f"Confidence: {np.max(probabilities):.2%}")
except Exception as e:
    print("Classifier needs training with real data before prediction.")
```
